### Creating Visualizations

Now that the data is ready, you can use the dataset to create a number of different visualizations. You do not need to make the visualizations perfect; this is just for practice.

#### Line Chart

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

plt.plot(x, y1, label='Sine')

plt.xlabel('X-axis')

plt.ylabel('Y-axis')
plt.title('Line Chart')
plt.legend()

plt.show()
```

```python
# Multiple Line chart 
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')

plt.xlabel('X-axis')

plt.ylabel('Y-axis')
plt.title('Multiple Line Chart')
plt.legend()

plt.show()
```

#### Bar Chart

```python
df['key'].value_counts().plot(kind='bar')
```

```python
# Horizontal bar chart
df['key'].value_counts().plot(kind='barh')
value_counts = df['key'].value_counts()

plt.barh(['key'], value_counts)

plt.xlabel('value_counts')
plt.ylabel('key')
plt.title('Horizontal Bar Chart')

plt.show()
```

```python
# Grouped bar chart
# Data for the two groups
group1 = [10, 20, 30]
group2 = [15, 25, 35]

# Combine the data into a list of lists
data = [group1, group2]

# Create labels for the groups and bars
labels = ['Group 1', 'Group 2']
bar_labels = ['A', 'B', 'C']

width = 0.35  # Set the width of the bars
index = np.arange(len(bar_labels))  # Create an index for the bars

plt.bar(index, group1, width, label=labels[0])
plt.bar(index + width, group2, width, label=labels[1])

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(index + width / 2, bar_labels)
plt.legend()

plt.show()
```
