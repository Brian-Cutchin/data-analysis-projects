Below we set up the dataframe with the Goodreads dataset for you. If you run this cell, you may get an error that some lines were skipped. Do not worry about that! For this studio, we want to focus on creating an effective visualization.

```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()

```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here. matplotlib Number of books published per year.
books_per_year = df.groupby('year').size()
#print(books_per_year)

plt.scatter(books_per_year.index, books_per_year.values)

plt.xlabel('Year')
plt.ylabel('Number of Books Published')
plt.title('Books Published per Year')
plt.xticks(rotation=90)  # Rotate labels by 90 degrees

# Show the plot
plt.show()
```

```python
# Use your Python skills to create a visualization here. seaborn Number of books published per year.

# Group by 'Year' and count the number of books
books_per_year = df.groupby('year').size()

# Create a line plot using Seaborn
sb.scatterplot(x=books_per_year.index, y=books_per_year.values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Books Published')
plt.title('Books Published per Year')
plt.xticks(rotation=90)  # Rotate labels by 90 degrees

# Show the plot
plt.show()
```
